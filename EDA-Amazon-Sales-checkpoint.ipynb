### Feature Engineer

```python
#Criaremos faixas de avaliação e de valor para cada instância da base de dados
def faixa_preco(x):
    for i in range(0, 1000, 100):
        if i < x <= i + 100:
            return f'{i+1}-{i+100}'
    return '1000+'

def faixa_estrela(x):
    for i in range(0, 5):
        if i < x <= i + 1:
            return f'{i}-{i+1}'
    return x

df_product['faixa_estrela'] = df_product['stars'].apply(faixa_estrela)
```

```python
#A base de dados possui total de 206 categorias de produtos
df_product['categoryName'].nunique()
```

```python
#Apresentamos os 10 produtos mais comprados na Amazon Brasil 2023
resumo_por_categoria = df_product.groupby('categoryName').agg({'price':['count','mean','median','sum']})

resumo_por_categoria.reset_index(inplace=True)
```

```python
resumo_por_categoria
```

```python
#Listamos os nomes que desejamos no agrupamento
resumo_por_categoria.columns = ['Categoria do produto','Quantidade vendida','Ticket Médio','Mediana','Montante']

resumo_por_categoria['Ticket Médio'] = round(resumo_por_categoria['Ticket Médio'], 2)
```
